```python
#Function to save the imported raw data file in the form of CSV file

def save_dataset(df, filename, file_format="csv"):
    if file_format == "csv":
        df.to_csv(filename, index=False)
    elif file_format == "json":
        df.to_json(filename, orient="records", lines=True)
    else:
        raise ValueError("Unsupported file format. Use 'json' or 'csv'.")
```

```python
# function to create prompt for training the model

def create_prompt(instruction, input_text, output_text):
    prompt = (
        f"<|begin_of_text|>"
        f"<|start_header_id|>system<|end_header_id|>\n"
        f"{instruction}\n"
        f"<|start_header_id|>user<|end_header_id|>\n"
        f"{input_text}\n"
        f"<|start_header_id|>assistant<|end_header_id|>\n"
        f"{output_text}\n"
        f"<|eot_id|>"
    )
    return prompt
```

```python
def process_dataset_for_training(dataset, output_file, variable_name):
    df = dataset

    # Ensure required columns exist
    required_columns = {"instruction", "input", "output"}
    if not required_columns.issubset(df.columns):
        raise ValueError(f"The dataset must contain the following columns: {required_columns} \nColumns:{df.columns} \nFile Name:{variable_name}")

    # Generate prompts
    with open(output_file, "a", encoding="utf-8") as f:
        for _, row in df.iterrows():
            instruction = row["instruction"]
            input_text = row["input"]
            output_text = row["output"]
            prompt = create_prompt(instruction, input_text, output_text)
            json.dump({"text": prompt}, f)
            f.write("\n")

    print(f"Processed {len(df)} prompts. Saved to {output_file}")
```

```python
dataset_path = "../data/processedData"  # Replace with your dataset file path
output_file = "../data/prompts.jsonl"  # Replace with your desired output file path

# process_dataset_for_training(dataset_file, output_file)

import os

# Read each CSV file and assign it to a variable
for file in os.listdir(dataset_path):
    if file.endswith(".csv"):
        filepath = os.path.join(dataset_path, file)
        variable_name = os.path.splitext(file)[0] + "_df"
        df = pd.read_csv(filepath)
        if len(df) > 50000:
            df = df.sample(n=30000, random_state=42)
            df.to_csv(f"{dataset_path}/{variable_name}_short.csv", index=False)

        if len(df) <= 50000:
            process_dataset_for_training(df, output_file, variable_name)
```
